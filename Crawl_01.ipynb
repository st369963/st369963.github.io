```python
import requests as rq
from bs4 import BeautifulSoup
import time
import pandas as pd
```

```python
title = []
body = []
publish_time = []
```

```python
home_url = "https://tw.news.yahoo.com/entertainment/archive"
r = rq.get(home_url)
soup = BeautifulSoup(r.text, "lxml")
print(soup)
```

```python
for i in range(0, len(soup.select("div.Cf"))):
    url = soup.select("div.Cf")[i].select("a")[0]["href"]
    title_time = soup.select("div.Cf")[i].select("a")[0].text
    
    r_content =rq.get(home_url + url)
    soup_content = BeautifulSoup(r_content.text,"lxml")
    title.append(soup_content.find("header", "caas-title-wrapper").text)
    body.append("".join(str(x) for x in soup_content.select("div.caas-body p")))
    publish_time.append(soup_content.find("div", "caas-attr-time-style").text)
    #爬太快容易被ban 設置一個停頓的時間
    print(soup_content.find("header", "caas-title-wrapper").text)
    time.sleep(1)
```

```python
df = pd.DataFrame({"title":title, "body":body, "publish_time":publish_time})
df
```
